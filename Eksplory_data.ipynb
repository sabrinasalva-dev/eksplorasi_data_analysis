4. Visualisasikan jumlah nilai unik

```python
#ini diambil dari beberapa data yang ada karena data tersebut variabel yang ingin kita visualisasikan. Variabel ini mencakup kategori dan metrik yang berbeda.
variables = ['university', 'degree_level', 'cgpa', 'sports_engagement', 'average_sleep', 'academic_pressure'] #Kita mendefinisikan daftar variabel yang ingin kita visualisasikan. Variabel ini mencakup kategori dan metrik yang berbeda.



plt.figure(figsize=(12, 12)) #ni mengatur ukuran figure (plot) menjadi 12x12 inci.


for i, var in enumerate(variables): #Loop ini akan iterasi melalui setiap variabel dalam daftar variables. i adalah indeks dan var adalah nama variabel yang sedang diproses.

    plt.subplot(3, 2, i + 1)  #ni membuat subplot dalam grid 3x2 (3 baris dan 2 kolom). i + 1 memastikan kita menggunakan indeks yang benar untuk subplot.
    if data[var].dtype == 'object': 
        sns.countplot(data=data, x='gender', hue=var, palette='pastel') #Jika variabel bersifat kontinu (misalnya, cgpa atau average_sleep), kita menggunakan boxplot untuk menunjukkan distribusi nilai.
        plt.title(f'uiversity {var} by all') #kasih judul
        plt.title(f'Degre_level of {var} by all') #kasih judul
        plt.ylabel('Count') #memberikan range 
        plt.legend(title=var)
    else:  # Continuous
        sns.boxplot(data=data, x='gender', y=var, palette='pastel') #Jika variabel bersifat kontinu (misalnya, cgpa atau average_sleep), kita menggunakan boxplot untuk menunjukkan distribusi nilai
        plt.title(f'{var} Distribution by all')
        plt.ylabel(var) 

plt.tight_layout()  #Ini memastikan bahwa subplot tidak saling tumpang tindih dan semuanya terlihat rapi.
plt.show()  # Show plots
```

```python
if 'gender' in data.columns and 'anxiety' in data.columns: #Baris ini memeriksa apakah kolom gender dan anxiety ada dalam DataFrame data. Jika keduanya ada, maka kode di dalam blok if akan dijalankan.
    # Mengonversi 'gender' menjadi numerik
    data['gender_numeric'] = data['gender'].astype('category').cat.codes #Kolom gender, yang biasanya berupa teks (misalnya "Male" dan "Female"), diubah menjadi format numerik menggunakan metode .astype('category')

    # Menghapus nilai NaN
    x = data['gender_numeric'].dropna() #Baris ini menghapus nilai kosong (NaN) dari kolom gender_numeric dan anxiety. 
    y = data['anxiety'].dropna()

    # Membuat histogram dua dimensi
    plt.figure(figsize=(10, 6)) #Mempersiapkan figure untuk plot dengan ukuran 10x6 inci. Ini menentukan ukuran visualisasi yang akan ditampilkan.
    plt.hist2d(x, y, bins=(2, 20), cmap='Blues')  # Jumlah bin untuk gender dan anxiety
    plt.colorbar(label='Frekuensi') #Fungsi ini menambahkan color bar di samping histogram, yang menunjukkan seberapa banyak frekuensi untuk setiap kombinasi bin.
    plt.title('Histogram Dua Dimensi: Gender vs. Anxiety', fontsize=16) #Menetapkan judul untuk histogram dengan ukuran font 16. Judul ini menjelaskan isi plot.
    plt.xlabel('Gender (0=Female, 1=Male)', fontsize=14)  # Menampilkan label yang lebih informatif
    plt.ylabel('Anxiety', fontsize=14) #Memberikan label untuk sumbu x dan y dengan ukuran font 14. Label pada sumbu x menjelaskan bahwa nilai 0 mewakili perempuan dan 1 mewakili laki-laki.
    plt.grid(False) #Menghilangkan garis grid di latar belakang untuk tampilan yang lebih bersih.

    plt.show() #Fungsi ini menampilkan histogram yang telah dibuat. Jika tidak ada perintah ini, plot tidak akan muncul.
else:
    print("Kolom tidak ditemukan.")
```

9. create a box plot

```python
#ini kita ambil yang relavan

plt.figure(figsize=(15, 10)) ## Menentukan ukuran figure

# Membuat boxplot untuk kolom 'age'
plt.subplot(2, 2, 1)  # Baris 2, Kolom 2, Plot ke-1
sns.boxplot(y=data['age']) #ini adalah fungsi dari Seaborn yang digunakan untuk membuat boxplot
plt.title('Boxplot of Age')#Ini menambahkan judul pada plot untuk memberikan konteks kepada pembaca mengenai apa yang ditampilkan. Judulnya menyatakan bahwa plot tersebut menunjukkan boxplot untuk variabel age.

# Membuat boxplot untuk kolom 'cgpa'
plt.subplot(2, 2, 2)  # Baris 2, Kolom 2, Plot ke-2
sns.boxplot(y=data['cgpa']) 
plt.title('Boxplot of CGPA')

# Membuat boxplot untuk kolom 'depression'
plt.subplot(2, 2, 3)  # Baris 2, Kolom 2, Plot ke-3
sns.boxplot(y=data['depression'])
plt.title('Boxplot of Depression')

# Membuat boxplot untuk kolom 'anxiety'
plt.subplot(2, 2, 4)  # Baris 2, Kolom 2, Plot ke-4
sns.boxplot(y=data['anxiety'])
plt.title('Boxplot of Anxiety')

# Menampilkan plot
plt.tight_layout() #gunakan untuk mengatur tata letak subplot dalam figure.
plt.show() #digunakan untuk menampilkan semua plot yang telah Anda buat.
```

```python
data_cleaned = data[['anxiety', 'isolation', 'cgpa', 'academic_pressure', 'study_satisfaction', 'average_sleep']].dropna() # Baris ini memilih kolom yang relevan dari DataFrame data, yaitu anxiety, isolation, cgpa, academic_pressure, study_satisfaction, dan average_sleep.

# Menyiapkan figure untuk 4 scatter plot
plt.figure(figsize=(14, 10)) #Mempersiapkan figure dengan ukuran 14x10 inci, yang akan menjadi tempat untuk menampilkan scatter plot.


# Scatter plot 1: Anxiety vs. Isolation
plt.subplot(2, 2, 1) #mengatur layout untuk 4 plot dalam 2 baris dan 2 kolom, dan memilih subplot pertama.
plt.scatter(data_cleaned['anxiety'], data_cleaned['isolation'], alpha=0.7, color='blue', edgecolors='k') #membuat plot dengan anxiety pada sumbu x dan isolation pada sumbu y. 
plt.title('Scatter Plot: Anxiety vs. Isolation') #Menambahkan judul, label sumbu x dan y, serta mengaktifkan grid untuk memudahkan pembacaan.
plt.xlabel('Anxiety')
plt.ylabel('Isolation')
plt.grid(True)

# Scatter plot 2: CGPA vs. Academic Pressure
plt.subplot(2, 2, 2) #Mengatur subplot kedua untuk plot kedua.
plt.scatter(data_cleaned['cgpa'], data_cleaned['academic_pressure'], alpha=0.7, color='green', edgecolors='k') #Menampilkan hubungan antara cgpa dan academic_pressure.
plt.title('Scatter Plot: CGPA vs. Academic Pressure')
plt.xlabel('CGPA')
plt.ylabel('Academic Pressure')
plt.grid(True)

# Scatter plot 3: Study Satisfaction vs. Average Sleep
plt.subplot(2, 2, 3) #Mengatur subplot ketiga.
plt.scatter(data_cleaned['study_satisfaction'], data_cleaned['average_sleep'], alpha=0.7, color='orange', edgecolors='k')
plt.title('Scatter Plot: Study Satisfaction vs. Average Sleep')
plt.xlabel('Study Satisfaction')
plt.ylabel('Average Sleep')
plt.grid(True)

# Scatter plot 4: CGPA vs. Anxiety
plt.subplot(2, 2, 4)
plt.scatter(data_cleaned['cgpa'], data_cleaned['anxiety'], alpha=0.7, color='purple', edgecolors='k')
plt.title('Scatter Plot: CGPA vs. Anxiety')
plt.xlabel('CGPA')
plt.ylabel('Anxiety')
plt.grid(True)

# Menampilkan semua scatter plot
plt.tight_layout()
plt.show()
```

4. Visualisasikan jumlah nilai unik

```python
#ini diambil dari beberapa data yang ada karena

variables = ['university', 'degree_level', 'cgpa', 'sports_engagement', 'average_sleep', 'academic_pressure'] #Kita mendefinisikan daftar variabel yang ingin kita visualisasikan. Variabel ini mencakup kategori dan metrik yang berbeda.


plt.figure(figsize=(12, 12)) #ni mengatur ukuran figure (plot) menjadi 12x12 inci.


for i, var in enumerate(variables): #Loop ini akan iterasi melalui setiap variabel dalam daftar variables. i adalah indeks dan var adalah nama variabel yang sedang diproses.

    plt.subplot(3, 2, i + 1)  #ni membuat subplot dalam grid 3x2 (3 baris dan 2 kolom). i + 1 memastikan kita menggunakan indeks yang benar untuk subplot.
    if data[var].dtype == 'object': 
        sns.countplot(data=data, x='gender', hue=var, palette='pastel') #Jika variabel bersifat kontinu (misalnya, cgpa atau average_sleep), kita menggunakan boxplot untuk menunjukkan distribusi nilai.
        plt.title(f'uiversity {var} by all') #kasih judul
        plt.title(f'Degre_level of {var} by all') #kasih judul
        plt.ylabel('Count') #memberikan range 
        plt.legend(title=var)
    else:  # Continuous
        sns.boxplot(data=data, x='gender', y=var, palette='pastel') #Jika variabel bersifat kontinu (misalnya, cgpa atau average_sleep), kita menggunakan boxplot untuk menunjukkan distribusi nilai
        plt.title(f'{var} Distribution by all')
        plt.ylabel(var) 

plt.tight_layout()  #Ini memastikan bahwa subplot tidak saling tumpang tindih dan semuanya terlihat rapi.
plt.show()  # Show plots
```

```python
#ini adalah daftar nama-nama variabel yang akan dieksplorasi. Variabel ini mencakup baik yang bersifat kategorikal (seperti university dan degree_level) maupun yang bersifat kontinu (seperti cgpa, average_sleep, dan academic_pressure).
variables = ['university', 'degree_level', 'cgpa', 'sports_engagement', 'average_sleep', 'academic_pressure']

# Print nilai unik untuk setiap variabel
for var in variables: #Di sini, kita menggunakan loop for untuk iterasi melalui setiap variabel dalam daftar variables. Variabel saat ini dalam iterasi disimpan dalam var.
    unique_values = data[var].unique() #Metode .unique() dari Pandas digunakan untuk mengambil semua nilai unik yang ada dalam kolom var dari DataFrame data. 
    print(f'Nilai unik di {var}: {unique_values}') #Menggunakan f-string, kita mencetak nama variabel dan daftar nilai unik yang ditemukan. Ini memberi informasi langsung tentang variasi data untuk setiap kolom.

```

9. create a box plot

```python
#ini kita ambil yang relavan

plt.figure(figsize=(15, 10)) ## Menentukan ukuran figure

# Membuat boxplot untuk kolom 'age'
plt.subplot(2, 2, 1)  # Baris 2, Kolom 2, Plot ke-1
sns.boxplot(y=data['age']) #ini adalah fungsi dari Seaborn yang digunakan untuk membuat boxplot
plt.title('Boxplot of Age')#Ini menambahkan judul pada plot untuk memberikan konteks kepada pembaca mengenai apa yang ditampilkan. Judulnya menyatakan bahwa plot tersebut menunjukkan boxplot untuk variabel age.

# Membuat boxplot untuk kolom 'cgpa'
plt.subplot(2, 2, 2)  # Baris 2, Kolom 2, Plot ke-2
sns.boxplot(y=data['cgpa']) 
plt.title('Boxplot of CGPA')

# Membuat boxplot untuk kolom 'depression'
plt.subplot(2, 2, 3)  # Baris 2, Kolom 2, Plot ke-3
sns.boxplot(y=data['depression'])
plt.title('Boxplot of Depression')

# Membuat boxplot untuk kolom 'anxiety'
plt.subplot(2, 2, 4)  # Baris 2, Kolom 2, Plot ke-4
sns.boxplot(y=data['anxiety'])
plt.title('Boxplot of Anxiety')

# Menampilkan plot
plt.tight_layout() #gunakan untuk mengatur tata letak subplot dalam figure.
plt.show() #digunakan untuk menampilkan semua plot yang telah Anda buat.
```
